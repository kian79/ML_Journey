```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

```python
train_path = os.path.join('data','train.csv')
test_path = os.path.join('data','test.csv')
```

```python
train_df = pd.read_csv(train_path)
```

```python
test_df = pd.read_csv(test_path)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
def clean_data(input_df):
    df = input_df
    oe = OrdinalEncoder()
    df['MSZoning'] = oe.fit_transform(df[['MSZoning']])
    imputer = SimpleImputer(strategy="median")
    df['LotFrontage'] = imputer.fit_transform(df[['LotFrontage']])
    df.Street = oe.fit_transform(df[['Street']])
#     df.Alley = df.Alley.replace(np.nan, 'No', regex=True)
#     df.Alley = oe.fit_transform(df[['Alley']])
    df.LotShape = oe.fit_transform(df[['LotShape']])
    df[df.columns[8:17]] = oe.fit_transform(df[df.columns[8:17]])
    df[df.columns[21:25]] = oe.fit_transform(df[df.columns[21:25]])
    df.MasVnrType = df.MasVnrType.replace(np.nan, 'None', regex=True)
    df.MasVnrType = oe.fit_transform(df[['MasVnrType']])
    imputer = SimpleImputer(strategy="median")
    df['MasVnrArea'] = imputer.fit_transform(df[['MasVnrArea']])
    df.BsmtQual = df.BsmtQual.replace(np.nan, 'No', regex=True)
    df.BsmtCond = df.BsmtCond.replace(np.nan, 'No', regex=True)
    df.BsmtExposure = df.BsmtExposure.replace(np.nan, 'No', regex=True)
    df.BsmtFinType1 = df.BsmtFinType1.replace(np.nan, 'No', regex=True)
    df.BsmtFinType2 = df.BsmtFinType2.replace(np.nan, 'No', regex=True)
    df[df.columns[27:34]] = oe.fit_transform(df[df.columns[27:34]])
    df[['BsmtFinType2']] = oe.fit_transform(df[['BsmtFinType2']])
    df.Electrical = df.Electrical.replace(np.nan, 'SBrkr', regex=True)
    df[df.columns[39:43]] = oe.fit_transform(df[df.columns[39:43]])
    df[['KitchenQual']] = oe.fit_transform(df[['KitchenQual']])
    df[['Functional']] = oe.fit_transform(df[['Functional']])
    df = df.drop(columns=['PoolQC','Fence','MiscFeature','FireplaceQu'])
    df[df.columns[75:76]] = oe.fit_transform(df[df.columns[75:76]])
    df.GarageQual = df.GarageQual.replace(np.nan, 'No', regex=True)
    df.GarageCond = df.GarageCond.replace(np.nan, 'No', regex=True)
    df.GarageYrBlt = df.GarageYrBlt.replace('No', np.nan, regex=True)
    df.GarageFinish = df.GarageFinish.replace(np.nan, 'No', regex=True)
    df.GarageType = df.GarageType.replace(np.nan, 'No', regex=True)
    df[df.columns[62:65]] = oe.fit_transform(df[df.columns[62:65]])
    df[df.columns[57:58]] = oe.fit_transform(df[df.columns[57:58]])
    df[df.columns[59:60]] = oe.fit_transform(df[df.columns[59:60]])
    imputer = SimpleImputer(strategy="median")
    df['GarageYrBlt'] =imputer.fit_transform(df[['GarageYrBlt']])
    df.drop(['Alley'],axis=1,inplace=True)
    return df
```
